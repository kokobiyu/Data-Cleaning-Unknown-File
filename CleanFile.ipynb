```python
import pandas as pd
from io import StringIO
```

```python
def CleanFile(file):
    with open(file, "r") as f2:
        data = f2.read()
    
    special_characters=['@','#','$','*','&','!', '%', '\\']
    data = "".join(filter(lambda char: char not in special_characters , data))
    StringData = StringIO(data)
 
    # let's read the data using the Pandas read_csv() function
    df = pd.read_csv(StringData)
    df.columns = [' ']
    
    df2 = df[df[' '].str.endswith('A')].loc[:,' '].apply(lambda x: x[::-1])
    df2 = pd.DataFrame(df2)
    df.update(df2)
    
    df[[
    'Target Designation',
    'Target Race',
    'Target Priority',
    'Target Shape',
    'Target Length',
    'Target Width',
    'Number of Target Components',
    'Target Category',
    'Target Surveillance',
    'Target Type',
    'Target Density',
    'Target Mobility',
    'Target Protection',
    'Target X Coordinate',
    'Target Y Coordinate',
    'Target Observer',
    'Target Discovered Game Time',
    'Target Health',
    'Target Status',
    'Target Point Value']] = df[' '].str.split(' ', expand=True)
    df.drop([' '], axis=1, inplace=True)
    df.to_csv('clean_file.csv')
    return df
    
```

```python
CleanFile("SC2TGT_ListCorrupted")
```
