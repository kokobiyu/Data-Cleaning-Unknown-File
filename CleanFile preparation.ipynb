```python
import pandas as pd
from io import StringIO
```

```python
with open("SC2TGT_ListCorrupted", "r") as f2:
    data = f2.read()

# removing special character from text
special_characters=['@','#','$','*','&','!', '%', '\\']
data = "".join(filter(lambda char: char not in special_characters , data))
```

```python
StringData = StringIO(data)
 
# let's read the data using the Pandas read_csv() function
df = pd.read_csv(StringData)
```

```python
df.columns = [' ']
df.head(10)
```

```python
df[df[' '].str.endswith('A')]
```

looks like there are a lot of corrupted files where the texts are reversed

```python
# creating new dataframe of corrupted string and reverse them
df2 = df[df[' '].str.endswith('A')].loc[:,' '].apply(lambda x: x[::-1])
df2 = pd.DataFrame(df2)
df2
```

```python
df.update(df2) # update the dataframe
df
```

```python
df[[
'Target Designation',
'Target Race',
'Target Priority',
'Target Shape',
'Target Length',
'Target Width',
'Number of Target Components',
'Target Category',
'Target Surveillance',
'Target Type',
'Target Density',
'Target Mobility',
'Target Protection',
'Target X Coordinate',
'Target Y Coordinate',
'Target Observer',
'Target Discovered Game Time',
'Target Health',
'Target Status',
'Target Point Value']] = df[' '].str.split(' ', expand=True)
df.drop([' '], axis=1, inplace=True)
```
